### Introduccion

#### About

La <a href='https://es.wikipedia.org/wiki/Premier_League'>Premier League</a>, también conocida en Inglaterra como The Premiership, es la máxima categoría del sistema de ligas de fútbol de Inglaterra. Comenzó a disputarse en la temporada 1992-93. Es considerada una de las cinco grandes ligas europeas junto con la Primera División española, la Serie A italiana, la Bundesliga alemana y la Ligue 1 francesa, ocupando el primer puesto del coeficiente UEFA.

#### Dataset
El <a href='https://www.kaggle.com/rajatrc1705/english-premier-league202021'>dataset</a> está compuesto por los datos de la temporada 20-21 de la EFL, en cada fila los datos representan a un jugador de fútbol y las columnas se utilizan para describir los datos de los jugadores. En total tiene 532 filas x 18 columnas, las cuales algunas pueden ser:

* Nombre
* Club
* Nacionalidad
* Posicion
* Edad
* Goles
* Asistencias
* Etc.

### Análisis

#### Maximos Goleadores

```python
### Ordeno la data en forma descendente y que solo muestre los primeros 10 registros 
data_sort=data.sort_values('Goals', ascending=True)[-10:]

#Modificaciones
plt.title('Top Goleadores de la EFL 20-21',fontsize=16)
plt.xlabel('Goles')

# Grafico
plt.barh(data_sort['Name'],data_sort['Goals'], color='#54796d')
plt.grid()
plt.show()
```

#### Según los datos obtenidos en el grafico el goleador de la EFL 20-21 fue Harry Kane seguido por Mohamed Salah y Bruno Fernandes respectivamente.

#### Maximos Asistentes

```python
### Ordeno la data en forma descendente y que solo muestre los primeros 10 registros 
data_sort=data.sort_values('Assists', ascending=True)[-10:]

#Modificaciones
plt.title('Top Asistidores de la EFL 20-21',fontsize=16)
plt.xlabel('Asistencias')

# Grafico
plt.barh(data_sort['Name'],data_sort['Assists'])
plt.grid()
plt.show()
```

#### Según los datos obtenidos en el grafico el maximo asistidor de la EFL 20-21 fue Harry Kane seguido por Kevin de Bruyne y Bruno Fernandes respectivamente.

### Top de Jugadores Amonestados

```python
### Ordeno la data en forma descendente y que solo muestre los primeros 10 registros 
data_sort=data.sort_values('Yellow_Cards', ascending=True)[-10:]

#Modificaciones
plt.title('Top Amonestados de la EFL 20-21',fontsize=16)
plt.ylabel('Amonestaciones')
plt.xticks(rotation='vertical')

# Grafico
plt.bar(data_sort['Name'],data_sort['Yellow_Cards'], color='#FFDB58')
plt.show()
```

#### Según los datos obtenidos en el grafico John McGinn es el jugador con mas amonestaciones de la EFL 20-21, seguido por Harry Maguire y Connor Gallagher respectivamente.

### Top de Jugadores Expulsados

```python
### Ordeno la data en forma descendente y que solo muestre los primeros 10 registros 
data_sort=data.sort_values('Red_Cards', ascending=True)[-10:]

#Modificaciones
plt.title('Top Expulsados de la EFL 20-21',fontsize=16)
plt.ylabel('Expulsiones')
plt.xticks(rotation='vertical')

# Grafico
plt.bar(data_sort['Name'],data_sort['Red_Cards'], color='#DC143C')
plt.show()
```

#### Según los datos obtenidos en el grafico Lewis Dunk es el jugador con mas amonestaciones de la EFL 20-21.

### Densidad de edad en la Premier
